```python
import pandas as pd
```

```python
data = pd.read_csv('all_games.csv')
data.head()
```

```python
data = data.drop(['app_id', 'tags_str', 'header_image', 'genres_str'], axis=1)
data.head()
```

```python
import json
with open('steam_games_backup.json', 'r') as file:
  data = json.load(file)

df = pd.DataFrame.from_dict(data, orient='index')

df.head(5)
```

```python
len(df)
```

```python
df.isna().sum()
```

```python
df[df['name'] == 'Penumbra: Requiem']
```

```python
index_to_drop = df[df['name'] == 'Fog of War'].index[0]
df = df.drop(index_to_drop)
```

```python
df = df.drop(['short_description', 'header_image', 'screenshots', 'movies'], axis=1)
df.head()
```

```python
df['category_price'] = df['price'].apply(lambda x: 'Paid' if x > 0.0 else 'Free')
df.head()
```

```python
df['release_date'] = pd.to_datetime(df['release_date'])
df['release_year'] = df['release_date'].dt.year.astype(pd.Int64Dtype())

df.head()
```

```python
df[df['tags'].apply(lambda x: 'PvP' in x) == True]
```

```python
df.isna().sum()
```

```python
df[df['release_date'].isna()]
```

```python
df.to_csv('game_data.csv', index=False)
```

```python
print(df['release_year'].unique())
```

```python
count_df = df['release_year'].value_counts()
count_df = count_df.reset_index()
count_df.columns = ['release_year', 'count']
count_df = count_df.sort_values(by='release_year', ascending=True)
count_df
```

```python
count_df.to_csv('game_count.csv', index=False)
```
